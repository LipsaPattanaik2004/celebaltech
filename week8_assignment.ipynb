```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.docstore.document import Document

embed_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
documents = [Document(page_content=txt) for txt in docs]
vector_db = FAISS.from_documents(documents, embedding=embed_model)
```

```python
from langchain_openai import ChatOpenAI
from langchain.chains import RetrievalQA

llm = ChatOpenAI(model_name="gpt-3.5-turbo", openai_api_key="your-key-here")

rag_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=vector_db.as_retriever(search_kwargs={"k": 5}),
    chain_type="stuff"
)
```

```python
import pandas as pd

test_path = r"C:\Users\patta\Downloads\archive (4)\Test Dataset.csv"
df_test = pd.read_csv(test_path)

df_test = df_test.astype(str).fillna("Unknown")
```

```python
predictions = []

for row in df_test.itertuples():
    query = f"""
    Applicant ID: {row.Loan_ID}
    Gender: {row.Gender}, Married: {row.Married}, Dependents: {row.Dependents}
    Education: {row.Education}, Self_Employed: {row.Self_Employed}
    Income: ApplicantIncome={row.ApplicantIncome}, CoapplicantIncome={row.CoapplicantIncome}
    LoanAmount: {row.LoanAmount}, Term: {row.Loan_Amount_Term}
    Credit History: {row.Credit_History}, Property Area: {row.Property_Area}
    
    Should the loan be approved for this applicant?
    """
    
    response = rag_chain.run(query)
    predictions.append((row.Loan_ID, response))
```

```python
result_df = pd.DataFrame(predictions, columns=["Loan_ID", "Predicted_Answer"])
result_df.to_csv("loan_predictions_from_test.csv", index=False)
result_df.head()
```

```python
import pandas as pd

df_test = pd.read_csv(r"C:\Users\patta\Downloads\archive (4)\Test Dataset.csv")
df_test = df_test.astype(str).fillna("Unknown")

submission = []
for row in df_test.itertuples():
    query = f"""
    Applicant ID: {row.Loan_ID}
    Gender: {row.Gender}, Married: {row.Married}, Dependents: {row.Dependents}
    Education: {row.Education}, Self_Employed: {row.Self_Employed}
    Income: ApplicantIncome={row.ApplicantIncome}, CoapplicantIncome={row.CoapplicantIncome}
    LoanAmount: {row.LoanAmount}, Term: {row.Loan_Amount_Term}
    Credit History: {row.Credit_History}, Property Area: {row.Property_Area}
    
    Should the loan be approved for this applicant?
    """
    
    response = rag_chain.run(query).lower()  
    predicted = "Y" if "yes" in response or "approved" in response else "N"
    submission.append((row.Loan_ID, predicted))
```

```python
submission_df = pd.DataFrame(submission, columns=["Loan_ID", "Loan_Status"])
submission_df.to_csv("final_submission.csv", index=False)
submission_df.head()
```
